```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def paramitrized_G(a, l, n):
    def G(x, ksi, t):
        ret = 0
        for i in range(1, n):
            lam = np.pi * i / l
            ret += np.exp(-a**2 * lam**2 * t) * np.sin(x * lam) * np.sin(ksi * lam) / l        
        return ret
    return G
    
```

```python
G = paramitrized_G(1, 3*np.pi, 10000)
x = np.linspace(0, 3*np.pi, 100)
fih, ax = plt.subplots(nrows=1,  figsize=(10,5))
ax.grid()
y = G(x, 3, 1)
ax.plot(x, y)
plt.show()
```
