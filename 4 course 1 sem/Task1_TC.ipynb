```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from sympy import *
```

```python
u_zero = lambda x, th1, th2, th3: 0
u_control = lambda x, th1, th2, th3 : th1*x[0] + th2*x[1] + th3*x[2]

def lin_parametrized_system(a, u_func, th1, th2, th3):
    def lin_system(t, x):
        return np.array([
            x[1],
            x[0] + x[2],
            -x[1] - a* x[2] + u_func(x, th1, th2, th3),
        ])
    return lin_system

def parametrized_system(a, u_func, th1, th2, th3):
    def system(t, x):
        return np.array([
            x[1],
            0.5 * ((1 + x[2])**2 / (1 - x[0])**2 - 1),
            -(1 + x[2]) * x[1] / (1 - x[0]) - a * (1 - x[0]) * x[2] + (1 - x[0]) * u_func(x, th1, th2, th3),
        ])
    return system
```

```python
from scipy.integrate import solve_ivp

def plot_time_lin(system, tstr, init_cond):
    result = solve_ivp(system ,tstr, init_cond,
                   t_eval=np.linspace(tstr[0], tstr[1], 150))

    plt.plot(result.t, result.y[0], result.t, result.y[1], result.t, result.y[2])
    plt.legend([r'$x1$', r'$x2$', r'$x3$'], fontsize=20)
    plt.xlabel(r'time (s)')
    plt.title(r"$Oscillograms$", fontsize = 20)
    plt.grid()
    #plt.xlim(0.0, 20.0)
    fig = plt.gcf()
    fig.set_size_inches(10, 6)
    return fig, result
```
