```python
import numpy as np
import scipy as sp
from scipy.linalg import expm
import matplotlib.pyplot as plt
```

```python
h = 0.01

A = np.array([[0, 1, 0],
            [1, 0, 1],
            [0, -1, -5]])

B = np.array([0, 0, 1])
```

```python
def discrete_sys(A, B, h):
    lam, S = np.linalg.eig(A)
    invS = np.linalg.inv(S);
    diagExp = np.diag(np.exp(lam * h))
    
    Anew = np.dot(np.dot(S, diagExp), invS)
    
    diagInteg = diagExp - np.diag(np.ones(A.shape[0])) 
    
    for i in range(0, A.shape[0]):
        diagInteg[i][i] = diagInteg[i][i] / lam[i]
    
    Bnew = np.dot(np.dot(np.dot(S, diagInteg), invS) , B)
    
    return Anew, Bnew
    
```

```python
u_control = lambda x, th : th[0]*x[0] + th[1]*x[1] + th[2]*x[2]

def d_parametrized_system(A, B, u_func, th):
    def d_system(x):
        return np.dot(A, x) + B * u_func(x, th)
    return d_system
```

```python
def plot_graphs(system, num_it, h, init_cond, ax, title):
    result = np.array([init_cond])
    for i in range(1, num_it):
        ksi = system(result[i - 1])
        result = np.append(result, [ksi], axis=0)
    interval = np.arange(0, num_it) * h
    ax.plot(interval, result[:, 0], interval, result[:, 1], interval, result[:, 2])
    ax.legend([r'$ksi1$', r'$ksi2$', r'$ksi3$'], fontsize=20)
    ax.set_xlabel(r'time (s)')
    ax.set_title(title, fontsize = 15)
    ax.grid()
    return ax
```

```python
def control_matrix(A, B):
    ret = np.array(np.vstack(B))
    ret = np.hstack((ret, np.vstack(np.dot(A, B))))
    #ret = np.hstack((ret, np.vstack(np.dot(np.dot(A, A), B))))
    #ret = np.hstack((ret, np.vstack(np.dot(np.dot(np.dot(A, A), A), B))))
    for i in range(1, A.shape[0] - 1):
        ret = np.hstack((ret, np.vstack(np.dot(A, ret[:,i]))))
    return ret

def Frobenius_first(A, B):
    S1 = control_matrix(A, B)
    print ( "S1 =")
    print(S1)
    rank = np.linalg.matrix_rank(S1)
    print(rank)
    if (rank != S1.shape[0]):
        print ("SYSTEM NON-CONTROL, rank = ", rank)
        return None, None, None
    print ( "-S1 =")
    S1_inv = np.linalg.inv(S1)
    print(S1_inv)
    return np.dot(np.dot(S1_inv, A), S1), np.dot(S1_inv, B), S1

def Frobenius_second(A, B):
    _A, _B, S1 = Frobenius_first(A, B)
    print("Frobenius first A =")
    print(_A)
    print("B =")
    print(_B)
    _A = np.transpose(_A)
    _B[0] = 0
    _B[_B.shape[0] - 1] = 1
    S2 = control_matrix(_A, _B)
    print("S2 = ", S2)
    S2_inv = np.linalg.inv(S2)
    print("S2 inv = ", S2_inv)
    S = np.dot(S1, S2_inv)   
    return _A, _B, S

def find_th(A, B, desired_eigens):
    _A, _B, S = Frobenius_second(A, B)
    print("S ", S)
    desired_poly = np.poly((desired_eigens))
    
    tmp = (_A[_A.shape[0] - 1])[::-1]
    _th = (-tmp - desired_poly[1:])[::-1]
    print("TH_",_th)
    S_inv = np.linalg.inv(S)
    print("S_inv", S_inv)
    return np.dot(_th, S_inv)

```

```python
h = 0.01
Ad, Bd = discrete_sys(A, B, h)
print(Ad, np.round(Bd, 5))
lam, S = np.linalg.eig(Ad)
Ad = expm(A*h)
print(lam)

#invA = np.linalg.inv(A);
#Exp = Ad - np.diag(np.ones(A.shape[0]))
3#print(Exp)
#tmp = np.dot(Exp, B)
#Bd = np.dot(np.dot(invA, Exp), B)

print(Ad, Bd)
```

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=5, figsize=(10,25))
num_it = 500
init_cond = np.array([2., 2., 2])



f = d_parametrized_system(Ad, Bd, u_control, np.array([0, 0, 0]))

plot_graphs(d_parametrized_system(Ad, Bd, u_control, np.array([0, 0, 0])),
          num_it, h, init_cond,
          ax1, r"d system u=0")

desired_eigens = np.array([0.5 , 0.988680588347179, 0.953358080406381])
th = find_th(Ad, Bd, desired_eigens)


tmp = np.array([[th[0], th[1], th[2]], [th[0], th[1], th[2]], [th[0], th[1], th[2]]])
print(tmp)

plot_graphs(d_parametrized_system(Ad, Bd, u_control, th),
          num_it, h, init_cond,
          ax2, r"d system u=0")
```
