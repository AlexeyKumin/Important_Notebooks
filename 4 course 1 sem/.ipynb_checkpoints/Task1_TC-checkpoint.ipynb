```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
u_zero = lambda x, th1, th2, th3: 0

u_control = lambda x, th1, th2, th3 : th1*x[0] + th2*x[1] + th3*x[2]

def lin_parametrized_system(a, u_func, th1, th2, th3):
    def lin_system(t, x):
        return np.array([
                     x[1],
                     x[0] + x[2],
                     -x[1] - a* x[2] + u_func(x, th1, th2, th3)
                 ])
    return lin_system
    
def parametrized_system(a, u_func, th1, th2, th3):
    def system(t, x):
        return np.array([
                     x[1],
                     0.5 * ((1 + x[2])**2 / (1 - x[0])**2 - 1),
                     -(1 + x[2]) * x[1] / (1 + x[1]) - a * (1 - x[0]) * x[2] + (1 - x[0]) * u_func(x, th1, th2, th3)
                 ])
    return system
```

```python
from scipy.integrate import solve_ivp

def plot_time(system, tstr, init_cond, ax, title):
    result = solve_ivp(system ,tstr, init_cond, method='RK23',
    t_eval=np.linspace(tstr[0], tstr[1], 150))
    ax.plot(result.t, result.y[0], result.t, result.y[1], result.t, result.y[2])
    ax.legend([r'$x1$', r'$x2$', r'$x3$'], fontsize=20)
    ax.set_xlabel(r'time (s)')
    ax.set_title(title, fontsize = 15)
    ax.grid()
    return ax, result
```

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=5, figsize=(10,25))
t_interval = np.array([0, 15])
init_cond = np.array([2., 2., 2])

plot_time(lin_parametrized_system(5, u_control, 2, 2, 2),
          t_interval, init_cond,
          ax1, r"lin system u=0")
plot_time(lin_parametrized_system(5, u_control, -23.309, -21.422, -3.619),
          t_interval, init_cond,
          ax2, r"lin system a) th = (-23.309, -21.422, -3.619)")
plot_time(lin_parametrized_system(5, u_control, -15.826, -17.398, -3.081),
          t_interval, init_cond,
          ax3, r"lin system b) th = (-15.826, -17.398, -3.081)")
plot_time(lin_parametrized_system(5, u_control, -35.211, -21.968, -3.581),
          t_interval, init_cond,
          ax4, r"lin system v) th = (-35.211, -21.968, -3.581)")
plot_time(lin_parametrized_system(5, u_control, -4.872, -3.6, 1.1),
          t_interval, init_cond, 
          ax5, "lin system g) th = ( -4.872, -3.6, 1.1)")
plt.show()
```

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=5, figsize=(10,25))
t_interval = np.array([0, 20])
init_cond = np.array([0.2, 0.2, 0.2])

plot_time(parametrized_system(5, u_control, 2, 2, 2),
          t_interval, init_cond,
          ax1, r"system u=0")

plot_time(parametrized_system(5, u_control, -23.309, -21.422, -3.619),
          t_interval, init_cond,
          ax2, r"system a) th = (-23.309, -21.422, -3.619)")

plot_time(parametrized_system(5, u_control, -15.826, -17.398, -3.081),
          t_interval, init_cond,
          ax3, r"system b) th = (-15.826, -17.398, -3.081)")

plot_time(parametrized_system(5, u_control, -35.211, -21.968, -3.581),
          t_interval, init_cond,
          ax4, r"system v) th = (-35.211, -21.968, -3.581)")

plot_time(parametrized_system(5, u_control, -4.872, -3.6, 1.1),
          t_interval, init_cond,
          ax5, "system g) th = ( -4.872, -3.6, 1.1)")
plt.show()
```
