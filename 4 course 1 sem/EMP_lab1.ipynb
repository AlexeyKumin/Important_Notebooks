```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from sympy import *
```

```python
def tangens(x, l, a, b, g, d):
    return np.tan(x * l) - x * (a * d - g * b) / (a * g + b * x**2 *d)
```

```python
x = np.linspace(0, 5., 200)
f1 = tangens(x, 3, 1, 0, 1, 0)
f2 = x - x
plt.plot(x, f1)
plt.plot(x, f2)
#plt.legend([r'$\phi$'], fontsize=20)
#plt.xlabel(r'&lambda')
#plt.show()

idx = np.argwhere(np.diff(np.sign(f1 - f2))).flatten()
plt.plot(x[idx], f2[idx], 'ro')
```

```python
def xn4(x, l, n):    
    def la(n):
        return (np.pi*(2 * n - 1)/(2 * l))
    return sqrt(2. / l) * np.cos(x * la(n))

def xn5(x, l, n):    
    def la(n):
        return (np.pi * n / l)
    return sqrt(2. / l) * np.sin(x * la(n))
```

```python
x = np.linspace(0, 10., 100)
for i in range(1, 6):
    f = xn5(x, np.pi, i)
    plt.plot(x, f)
plt.legend([r'1', r'2',r'3', r'4',r'5'], fontsize=10)
```
